## making some charts

```python
import pandas as pd
import altair as alt
```

```python
df = pd.read_csv("michel-star-data.csv")
```

```python
df.head()
```

```python

#a chart with all the countries 

df_sorted = df.sort_values("Total", ascending=False)

base = alt.Chart(df_sorted)

bars = base.mark_bar().encode(
    x=alt.X("Country:N", sort="-y", title=None,
             axis=alt.Axis(labelAngle=-45, labelFontSize=11)),
    y=alt.Y("Total:Q", title="Total starred restaurants"),
    color=alt.condition(
        alt.datum.Country == "France",
        alt.value("#d8d0c4"),
        alt.value("#d8d0c4")
    ),
    tooltip=[
        alt.Tooltip("Country:N"),
        alt.Tooltip("Total:Q", title="Total stars"),
        alt.Tooltip("1 Star:Q"),
        alt.Tooltip("2 Stars:Q"),
        alt.Tooltip("3 Stars:Q"),
    ]
)

label = base.mark_text(
    dy=-12, fontSize=13, fontWeight="bold", color="#c0392b"
).encode(
    x=alt.X("Country:N", sort="-y"),
    y=alt.Y("Total:Q"),
    text=alt.condition(
        alt.datum.Country == "France",
        alt.value("France · 629"),
        alt.value("")
    )
)

chart = (bars + label).properties(
    width=700, height=380,
    title=alt.TitleParams(
        "France leads the world — by a wide margin",
        fontSize=16, fontWeight="bold", anchor="start", dy=-10
    )
).configure_view(
    strokeWidth=0
).configure_axis(
    grid=False,
    domainColor="#cccccc",
    tickColor="#cccccc",
    labelFont="Georgia, serif",
    titleFont="Georgia, serif",
    titleFontSize=12,
    titleColor="#555555"
).configure_title(
    font="Georgia, serif"
)

chart.show()
```
